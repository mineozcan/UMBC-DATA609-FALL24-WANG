## Step 4 - Categorize Each Country

Create a new column "Group". If a country is one of the seven G7 countried, the value of "Group" would be "G7", if a country of one of the five BRICS countries, the value of "Group" would be "BRICS". Otherwise, the value of "Group" would be "REST of WORLD".

```python


G7_COUNTRIES = ["USA", "CAN", "GBR", "FRA", "DEU", "ITA", "JPN"]
BRICS_ORIGINAL = ["BRA", "RUS", "IND", "CHN", "ZAF"]
BRICS_ALL_MEMBERS = ["BRA", "RUS", "IND", "CHN", "EGY", "ETH", "IRN", "SAU", "ZAF"]


g7_countries = data[data['Country Code'].str.strip().str.upper().isin(G7_COUNTRIES)]
g7_countries = g7_countries.copy()
g7_countries.loc[:, 'Group'] = "G7"
print("G7 Countries:")
display(g7_countries[['Country Name', 'Country Code', 'Group']])

brics_original_countries = data[data['Country Code'].str.strip().str.upper().isin(BRICS_ORIGINAL)]
brics_original_countries = brics_original_countries.copy()
brics_original_countries.loc[:, 'Group'] = "BRICS - Original"
print("BRICS - Original Countries:")
display(brics_original_countries[['Country Name', 'Country Code', 'Group']])


brics_all_members_countries = data[data['Country Code'].str.strip().str.upper().isin(BRICS_ALL_MEMBERS)]
brics_all_members_countries = brics_all_members_countries.copy()
brics_all_members_countries.loc[:, 'Group'] = "BRICS - All Members"
print("BRICS - All Members Countries:")
display(brics_all_members_countries[['Country Name', 'Country Code', 'Group']])


rest_of_world_countries = data[
    ~data['Country Code'].str.strip().str.upper().isin(G7_COUNTRIES + BRICS_ALL_MEMBERS)
]
rest_of_world_countries = rest_of_world_countries.copy()
rest_of_world_countries.loc[:, 'Group'] = "REST OF WORLD"
print("REST OF WORLD Countries:")
display(rest_of_world_countries[['Country Name', 'Country Code', 'Group']])

```
