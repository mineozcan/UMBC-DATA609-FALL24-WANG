## Step 4 - Categorize Each Country

Create a new column "Group". If a country is one of the seven G7 countried, the value of "Group" would be "G7", if a country of one of the five BRICS countries, the value of "Group" would be "BRICS". Otherwise, the value of "Group" would be "REST of WORLD".

```python
G7_COUNTRIES = ["USA", "CAN", "GBR", "FRA", "DEU", "ITA", "JPN"]
BRICS_COUNTRIES = ["BRA", "RUS", "IND", "CHN", "ZAF"]


def assign_group(row):
    if row['Country Code'] in G7_COUNTRIES:
        return "G7"
    elif row['Country Code'] in BRICS_COUNTRIES:
        return "BRICS"
    else:
        return "REST OF WORLD"


data['Group'] = data.apply(assign_group, axis=1)

display(data[['Country Name', 'Group']].drop_duplicates().reset_index(drop=True))


print("G7 Countries:")
display(data[data['Group'] == 'G7'][['Country Name', 'Group']].drop_duplicates().reset_index(drop=True))

print("\nBRICS Countries:")
display(data[data['Group'] == 'BRICS'][['Country Name', 'Group']].drop_duplicates().reset_index(drop=True))
```
