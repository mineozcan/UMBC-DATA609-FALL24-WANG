## HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python


print(df3.columns)

df_selected = df3[df3['Year']==2018][['Country', 'GDP per capita (current US$)', 'Life expectancy at birth, total (years)']]
df_selected = df_selected.rename(columns={"Country": "Country Name",
                                          "GDP per capita (current US$)": "GDP per Capita (current US$)",
                                          "Life expectancy at birth, total (years)": "Life Expectancy (years)"})

df_selected = df_selected.dropna()

fig = px.scatter(
                df_selected,
                x='GDP per Capita (current US$)',
                y='Life Expectancy (years)',
                text="Country Name",
                title=f'Wealth vs Health (Year: 2018)',
                labels={'GDP per Capita (current US$)': 'GDP per Capita (USD)',
                        'Life Expectancy (years)': 'Life Expectancy (Years)'},
                template="plotly_white"
)

fig.update_traces(marker=dict(size=4, opacity=0.7), selector=dict(mode='markers'))
fig.update_layout(autosize=False, width=3000, height=1000)

fig.show()


```

## Homework Question 3

Select a year of data. calculate total population of each region and make a pie chart to display the distribution of total population of all regions.

```python
selected_year = 2013
df_2013 = df3[df3['Year'] == selected_year]


df_population_by_region = df_2013.groupby("Region")["Population, total"].sum().reset_index()


df_population_by_region = df_population_by_region.rename(columns={"Population, total": "Total Population"})


fig = px.pie(
    df_population_by_region,
    names="Region",
    values="Total Population",
    title=f"Distribution of Total Population by Region in {selected_year}",
    template="plotly_white",  # Optional theme
    hole=0.3  # Creates a donut-like chart (optional)
)

# Show the plot
fig.show()
```
